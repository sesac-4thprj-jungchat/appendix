```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('./keywords_progress (1).csv')

df.head(11)
```

```python
from collections import Counter

all_keywords = []

for keyword_string in df['keywords'].dropna():
    keywords = [kw.strip() for kw in keyword_string.split(',') if kw.strip()]
    all_keywords.extend(keywords)

# 키워드 등장 횟수 세기
keyword_counts = Counter(all_keywords)

# Top 50 추출
top_50_keywords = keyword_counts.most_common(50)

# 결과 출력
for rank, (keyword, count) in enumerate(top_50_keywords, start=1):
    print(f"{rank}. {keyword} - {count}회")
```

```python
from collections import Counter

# 키워드 전체 수집
all_keywords = []

for keyword_string in df['keywords'].dropna():
    keywords = [kw.strip() for kw in keyword_string.split(',') if kw.strip()]
    all_keywords.extend(keywords)

# 키워드 빈도 계산
keyword_counts = Counter(all_keywords)
top_50_keywords = keyword_counts.most_common(50)

# txt 파일로 저장
with open("top_50_keywords.txt", "w", encoding="utf-8") as f:
    for rank, (keyword, count) in enumerate(top_50_keywords, start=1):
        f.write(f"{rank}. {keyword} - {count}회\n")

print("✅ top_50_keywords.txt 파일 저장 완료!")
```

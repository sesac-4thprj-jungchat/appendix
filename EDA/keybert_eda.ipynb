```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('./keybert_keywords.csv')
df.head()
```

```python
with open("./stopwords-ko.txt", "r", encoding="utf-8") as f:
    korean_stopwords = set([line.strip() for line in f if line.strip()])
```

```python
def clean_keywords(keyword_str, stopwords):
    keywords = [kw.strip() for kw in keyword_str.split(",")]
    cleaned = [kw for kw in keywords if all(sw not in kw for sw in stopwords)]
    return ", ".join(cleaned)
```

```python
df["keywords_cleaned"] = df["keywords"].apply(lambda x: clean_keywords(x, korean_stopwords))
```
